```python
#Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np 

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of plots

# Read in the data

df = pd.read_csv(r'C:\Users\andre\Downloads\movies.csv')
```

```python
# Looking at Data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
# Data types for our columns
df.dtypes
```

```python
#Remove missing value data
df = df.dropna()
```

```python
# Change Data type

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace = False, ascending=False).head()
```

```python
pd.set_option('display.max_rows', None) #shows all data
```

```python
#Drop duplicates

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
df.head()
```

```python
# Look at correlations
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')


plt.show
```

```python
# Plot Budget vs Gross with Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "yellow"})
```

```python
df.sort_values(by=['gross'], inplace = False, ascending=False).head()
```

```python
# Correlation within dataframe

df.corr(numeric_only=True)
```

```python
#plot correlation

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
df.head()
```

```python
df_numerated = df

for col_name in df_numerated.columns:
    if(df_numerated[col_name].dtype == 'object'):
        df_numerated[col_name] = df_numerated[col_name].astype('category')
        df_numerated[col_name] = df_numerated[col_name].cat.codes

df_numerated.head()
```

```python
#plot all data correlation

correlation_matrix_all = df_numerated.corr(numeric_only=True)

sns.heatmap(correlation_matrix_all, annot=True)

plt.title('Correlation Matrix All Data')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
#using unstack to see correlations faster

correlation_mat = df_numerated.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
# using sorted pairs to view correlations

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# view select field of sorted pairs

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr.head()
```
